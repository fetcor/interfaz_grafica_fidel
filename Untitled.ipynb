```python
import tkinter as tk

def interfaz():
   
    ventana = tk.Tk()
    ventana.title("calculadora")
    ventana.geometry("800x800+700+100")
    ventana.minsize(width=400, height=500)
    ventana.configure(bg="lightblue")
    
    
    def calculate_grade():
        try:
            successes = int(input_successes.get())

            grade = (successes*(0.25))-(40-successes)*0.10
            grade_label.config(text = grade)

        except ValueError:
            if input_successes == "":
                grade_label.config(text = "¿dato?")

            else:
                grade_label.config(text = "Error")

        finally:
            input_successes.config(state= "disabled")
            button_calculate_grade.config(state = "disabled")

        
    def delete():
        input_successes.config(state ="normal")
        button_calculate_grade.config(State= "normal")
        input_successes.delete(0,tk.END)
        button_calculate_grade.delete(0,tk.END)
        grade_label.config(text ="")
        
        
        
    label_title = tk.Label(ventana,
                           text = "calculate_grade_1",
                           bg = "lightblue", fg = "black",
                           font = ("consolas",20,"bold"),
                           relief = tk.GROOVE, bd =2,
                           padx = 20, pady =20)
    label_title.pack(padx =20, pady = 20)
    
    #frame first
    
    chart1 = tk.Frame(ventana,
                      bg = "lightblue")
    
    label_sb1 = tk.Label(chart1, text = "input_successes",
                         bg = "lightblue",
                         font = ("consolas", 10, "bold"),
                         width = 40,
                         anchor = tk.W)
    label_sb1.pack(side = tk.LEFT, padx =20, pady = 20)
    
    input_successes = tk.Entry(chart1,
                               bg = "white", fg = "black",
                               font = ("consolas",18, "bold"),
                               relief=tk.SUNKEN, bd=3,
                               width=10,
                               justify=tk.RIGHT,
                               state="normal")
    input_successes.pack(side=tk.LEFT, padx=20,pady=20)
    
    chart1.pack()
    
    #frame two
    
    
    chart2 = tk.Frame(ventana,bg ="lightblue")
    
    grade_label1= tk.Label(chart2,
                           text= "calificacion:",
                           bg="lightblue",
                           font=("consolas",10,"bold"),
                           width=22,
                           anchor=tk.W)
    grade_label1.pack(side=tk.LEFT, padx=20, pady=20)
    
    grade_label = tk.Label(chart2, text= " ",
                           bg = "lightgreen",
                           font=("consolas",18,"bold"),
                           relief = tk.GROOVE,
                           width=20,
                           anchor=tk.E)
    grade_label.pack(side=tk.LEFT,padx=20,  pady=20)
    
    chart2.pack()
    
    #frame three
    
    
    chart3= tk.Frame(ventana, bg ="lightblue")
    
    button_delete = tk.Button(chart3, 
                              text = "delete",
                              bg= "grey",
                              font=("consola",18,"bold"),
                              width=10,
                              command=delete)
    button_delete.pack(side=tk.LEFT, padx=20,pady=20)
    
    button_calculate_grade = tk.Button(chart3, 
                                       text= "calculate_grade",
                                       bg="orange",
                                       font=("consola",18, "bold"),
                                       width=10,
                                       state="normal",
                                       command=calculate_grade)
    
    button_calculate_grade.pack(side=tk.LEFT , padx=20, pady=20)
    
    chart3.pack()
    
    ventana.mainloop()


a = interfaz()





    
```
