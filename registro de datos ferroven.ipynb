```python
import tkinter as tk
import sqlite3

def Crea_la_conexión_a_la_base_de_datos():
    
    conexion = sqlite3.connect("empleados.db")

    # Crea el cursor
    cursor = conexion.cursor()

def Crea_la_tabla_de_empleados():
    cursor.execute("""
    CREATE TABLE empleados (
      nombre TEXT NOT NULL,
      apellido TEXT NOT NULL,
      dni TEXT NOT NULL,
      cargo TEXT NOT NULL
    )
    """)

# Registra los datos del empleado
def registrar_datos():
  # Obtiene los datos del empleado
  nombre = entry_nombre.get()
  apellido = entry_apellido.get()
  dni = entry_dni.get()
  cargo = entry_cargo.get()

  # Inserta los datos en la base de datos
  cursor.execute("INSERT INTO empleados (nombre, apellido, dni, cargo) VALUES (?, ?, ?, ?)", (nombre, apellido, dni, cargo))

  # Actualiza la base de datos
  conexion.commit()

  # Limpia los cuadros de texto
  entry_nombre.delete(0, "end")
  entry_apellido.delete(0, "end")
  entry_dni.delete(0, "end")
  entry_cargo.delete(0, "end")

# Llama a la función para registrar los datos
#registrar_datos()

def Cierra_la_conexión_a_la_base_de_datos():
    conexion.close()



# Crea la ventana principal
root = tk.Tk()

# Crea la etiqueta para el título
label_titulo = tk.Label(root, text="Registro de personal")

# Crea los cuadros de texto
entry_nombre = tk.Entry(root)
entry_apellido = tk.Entry(root)
entry_dni = tk.Entry(root)
entry_cargo = tk.Entry(root)

# Crea el botón para registrar los datos
button_registrar = tk.Button(root, text="Registrar", command=registrar_datos)

# Coloca los elementos en la ventana
label_titulo.pack()
entry_nombre.pack()
entry_apellido.pack()
entry_dni.pack()
entry_cargo.pack()
button_registrar.pack()

# Inicia la ventana principal
root.mainloop()


```
